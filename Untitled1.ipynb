```python
import pandas as pd
```

```python
data2021 = pd.read_excel('Rental-bond-lodgements-year-2021.xlsx', skiprows = 2)
```

```python
data_2022 = pd.read_excel('RentalBond_Lodgements_Year_2022.xlsx', skiprows =2)
```

```python
data_2023 = pd.read_excel('RentalBond_Lodgements_Year_2023.xlsx', skiprows = 2)
```

```python
data_2024 = pd.read_excel('rental-bond-lodgements-year-2024_1 (1).xlsx', skiprows = 2)
```

```python
df = pd.concat([data2021, data_2022,data_2023,data_2024 ])
```

```python
df
```

```python
df.info()
```

```python
df['Postcode'] = df['Postcode'].astype(int)
```

```python
df[df['Weekly Rent'] == 'U']
```

As we are doing rental analysis, the rent value is required. So, we remove all the unknown values in the weekly rent column

```python
filtered_df = df[df['Weekly Rent'] != 'U']
```

```python
filtered_df
```

```python
filtered_df.loc[:,'Weekly Rent'] = filtered_df.loc[:,'Weekly Rent'].astype(float)
```

```python
filtered_df
```

```python
filtered_df['Bedrooms'].value_counts()
```

As we have 1.2 million rows data so removing the unknown bedrooms will not change the fact of the analysis. So, we will be romoving the rows with the unkown bedrooms.

```python
Cleaned_df = filtered_df[filtered_df['Bedrooms'] != 'U']
```

```python
Cleaned_df
```

```python
Cleaned_df.loc[:,'Bedrooms'] = Cleaned_df.loc[:,'Bedrooms'].astype(int)
```

```python
Cleaned_df.info()
```

```python
Cleaned_df.to_csv("cleaned_rentaldata.csv", index=False, encoding='utf-8')
```

```python
Cleaned_df.loc[:,'Pricing_segment'] = Cleaned_df.loc[:,'Weekly Rent'].apply(
    lambda x: 'expensive' if x > 900 else ('affordable' if x > 600 else 'Inexpensive')
)
```

```python
Cleaned_df
```

```python
Cleaned_df['Pricing_segment'].value_counts()
```

```python
def segment_postcode(postcode):
    if 2000 <= postcode <= 2234:
        return 'Sydney Metro'
    elif 2250 <= postcode <= 2490:
        return 'Regional NSW'
    elif 2619 <= postcode <= 2898 or 2921 <= postcode <= 2999:
        return 'Remote NSW'
    else:
        return 'Other NSW'
    
```

```python
Cleaned_df.loc[:,'Postcode'].apply(segment_postcode)
```
